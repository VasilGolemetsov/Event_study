```python


start_date = "01-01-2015"
end_date = "01-01-2016"
ticker = tickers2[tickers2['Sector']==sectors[1]]['Ticker']
returns = web.get_data_yahoo(ticker, start_date, end_date, interval = "d")["Close"]
market_returns = web.get_data_yahoo("^IXIC", start_date, end_date, interval = 'd').pct_change()[1:]['Close']
risk_free_returns = web.get_data_yahoo('^TNX', start_date, end_date, interval = 'd').pct_change()[1:]['Close']
# returns = pd.DataFrame(returns).pct_change()[1:]
returns = returns.drop("AAF.L", axis=1).pct_change()[1:]



```

```python

# returns = returns.columns.droplevel(-1)

returns.columns = list(returns.columns)
returns.index.name = None
# returns = results.join()
market_returns = pd.DataFrame(market_returns)
market_returns.index.name = None
market_returns.columns = ["^IXIC"]
risk_free_returns = pd.DataFrame(risk_free_returns)
risk_free_returns.index.name = None
risk_free_returns.columns = ["^TNX"]
returns = returns.join(risk_free_returns["^TNX"])
returns = returns.join(market_returns["^IXIC"])

for i in list(returns.columns):
    returns = returns.astype({i:'float'})

for i in range(len(returns.columns[:-2])):
    returns[returns.columns[i]] = returns[returns.columns[i]] - returns['^TNX'].mean()
returns.dropna()


```

```python

returns = returns.dropna()
x = returns['^IXIC']
x_sm = sm.add_constant(x)
values = pd.DataFrame(columns=['0','1'])
for i in range(20):
    returns = returns.dropna()
    y = returns[returns.columns[i]]
    model = sm.OLS(y,x_sm)
    results = model.fit()
    dict = {'0':results.params[0], '1':results.params[1] }
    values = pd.concat([values,pd.DataFrame(dict,index= [i])])
```
